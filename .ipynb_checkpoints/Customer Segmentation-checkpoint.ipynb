## RFM Analysis
A simple yet powerful marketing technique is an analysis utilizing recency (how recent was the customer's last purchase), frequency (how often did the customer make a purchase in a given period) and monetary (how much money did the customer spend in a given period) data to identify the best customers and perform targeted marketing campaigns.

### Recency (R)
To calculate recency, we need to find out most recent purchase date of each customer and see how many days they are inactive for. After having no. of inactive days for each customer, we will apply K-means* clustering to assign customers a recency score.

Lets go ahead and calculate that.

```python
df_rfm = df.copy()
```

```python
today = pd.to_datetime("now")
```

```python
df_rfm['days_from_last_purchase'] = (today-df_rfm['trans_date']).astype('timedelta64[D]')
df_rfm.head()
```

```python
lastpurchase_day = df_rfm[['customer_id','days_from_last_purchase']].groupby('customer_id').min().reset_index()
```

```python
lastpurchase_day.rename(columns={'days_from_last_purchase':'recency'}, inplace=True)
```

```python
lastpurchase_day.head()
```

```python
#plot a recency histogram
plot_data = [go.Histogram(x=lastpurchase_day['recency'])]
plot_layout = go.Layout(title='Recency')
fig = go.Figure(data=plot_data, layout=plot_layout)

pyoff.iplot(fig)
```

### Assigning a recency score
We are going to apply K-means clustering to assign a recency score. But we should tell how many clusters we need to K-means algorithm. To find it out, we will apply Elbow Method. Elbow Method simply tells the optimal cluster number for optimal inertia.

```python
from sklearn.cluster import KMeans

sse={} # error
tx_recency = lastpurchase_day[['recency']]
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000).fit(tx_recency)
    tx_recency["clusters"] = kmeans.labels_  #cluster names corresponding to recency values 
    sse[k] = kmeans.inertia_ #sse corresponding to clusters
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.show()
```

Here it looks like 4 is the optimal one. Based on business requirements, we can go ahead with less or more clusters. We will be selecting 4 for this project

```python
#build 4 clusters for recency and add it to dataframe
kmeans = KMeans(n_clusters=4)
lastpurchase_day['recency_cluster'] = kmeans.fit_predict(lastpurchase_day[['recency']])
```

```python
lastpurchase_day.head()
```

```python
lastpurchase_day.groupby('recency_cluster')['recency'].describe()
```

### Ordering clusters

We have a cluster corresponding to each customer_id. But each cluster is randomly assigned. Cluster 2 is not better than cluster 1 for e.g. and so on. We want to give clusters according to most recent transactions.

We will first find the mean of recency value corresponding to each cluster. Then we will sort these values. Let's say cluster 0 has the most recent transactions mean value. From the above table we see that cluster 3 (mean recency 926) > cluster 1 > cluster 2 > cluster 0. That means that cluster 3 is most inactive and cluster 0 is most recent. We will give indices to these clusters as 0,1,2,3. So cluster 3 becomes cluster 0, cluster 1 becomes cluster 1, cluster 2 becomes cluster 2 and so on. Now we will drop the original cluster numbers and replace them with 0,1,2,3.

```python
#function for ordering cluster numbers
def order_cluster(cluster_field_name, target_field_name,df,ascending):
    new_cluster_field_name = 'new_' + cluster_field_name
    df_new = df.groupby(cluster_field_name)[target_field_name].mean().reset_index()
    df_new = df_new.sort_values(by=target_field_name,ascending=ascending).reset_index(drop=True)
    df_new['index'] = df_new.index
    df_final = pd.merge(df,df_new[[cluster_field_name,'index']], on=cluster_field_name)
    df_final = df_final.drop([cluster_field_name],axis=1)
    df_final = df_final.rename(columns={"index":cluster_field_name})
    return df_final

lastpurchase_day = order_cluster('recency_cluster', 'recency',lastpurchase_day,False)
```

```python
lastpurchase_day.groupby('recency_cluster')['recency'].describe()
```

```python
lastpurchase_day.head()
```

Great! cluster 1 earlier is now cluster0, cluster 2 earlier is now cluster 1 and so on. The clusters are arranged according to inactiviuty. Cluster 0 now is most inactive, cluster 3 is most active.

### Frequency (F)

To create frequency clusters, we need to find total number orders for each customer. First calculate this and see how frequency look like.

```python
freq = df_rfm[['customer_id','trans_date']].groupby(['customer_id']).count().reset_index()
freq.rename(columns={'trans_date':'frequency'}, inplace=True)

freq.head()
```

```python
#plot the histogram
plot_data = [go.Histogram(x=freq['frequency'])]
plot_layout = go.Layout(title='Frequency')
fig = go.Figure(data=plot_data, layout=plot_layout)

pyoff.iplot(fig)
```

### Frequency clusters
Determine the right number of clusters for K-Means by elbow method

```python
from sklearn.cluster import KMeans

sse={} # error
tx_recency = freq[['frequency']]
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000).fit(tx_recency)
    tx_recency["clusters"] = kmeans.labels_  #cluster names corresponding to recency values 
    sse[k] = kmeans.inertia_ #sse corresponding to clusters
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.show()
```

By Elbow method, clusters number should be 4 as after 4, the graph goes down.

```python
# Applying k-Means
kmeans=KMeans(n_clusters=4)
freq['frequency_cluster']=kmeans.fit_predict(freq[['frequency']])

#order the frequency cluster
freq = order_cluster('frequency_cluster', 'frequency', freq, True )
freq.groupby('frequency_cluster')['frequency'].describe()
```

### Monetary (M)
Let’s see how our customer database looks like when we cluster them based on monetary. We will calculate transaction amount for each customer, plot a histogram and apply the same clustering method.

```python
money = df_rfm[['customer_id','trans_amount']].groupby('customer_id').sum().reset_index()
money.rename(columns = {'trans_amount':'monetary'}, inplace = True)

money.head()
```

```python
#plot a recency histogram
plot_data = [go.Histogram(x=money['monetary'])]
plot_layout = go.Layout(title='Monetary')
fig = go.Figure(data=plot_data, layout=plot_layout)

pyoff.iplot(fig)
```

### Monetary clusters
Determine the right number of clusters for K-Means by elbow method just like before

```python
from sklearn.cluster import KMeans

sse={} # error
tx_recency = money[['monetary']]
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000).fit(tx_recency)
    tx_recency["clusters"] = kmeans.labels_  #cluster names corresponding to recency values 
    sse[k] = kmeans.inertia_ #sse corresponding to clusters
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.show()
```

From elbow's method, we find that clusters can be 3 or 4. Lets take 4 as the number of clusters

```python
#apply clustering
kmeans = KMeans(n_clusters=4)
money['monetary_cluster'] = kmeans.fit_predict(money[['monetary']])

#order the cluster numbers
money = order_cluster('monetary_cluster', 'monetary',money,True)

#show details of the dataframe
money.groupby('monetary_cluster')['monetary'].describe()
```

### RFM Score
After having RFM information for each customer, We will give each customer a RFM score based on his/her RFM data.

```python
# Merge the table first
rfm_data = pd.merge(lastpurchase_day, freq, on='customer_id')
```

```python
rfm_data = pd.merge(rfm_data, money, on='customer_id')
```

```python
rfm_data.head()
```

```python
#calculate rfm score and use mean() to see details
rfm_data['rfm_score'] = rfm_data['recency_cluster'] + rfm_data['frequency_cluster'] + rfm_data['monetary_cluster']
rfm_data.groupby('rfm_score')['recency','frequency','monetary'].mean()
```

Score 9 is our best customer, score 0 is our worst customer.

```python
rfm_data['segment'] = 'Low-Value'
rfm_data.loc[rfm_data['rfm_score']>3,'segment'] = 'Mid-Value' 
rfm_data.loc[rfm_data['rfm_score']>6,'segment'] = 'High-Value' 
```

```python
rfm_data
```

```python
rfm_data.groupby('segment').count()['rfm_score']
```

## Conclusions

After applying unsupervised machine learning using K-Means Clustering to identify different groups (clusters) for each customer, there are 3 main groups as explained below.
- Low Value: Customers who are less active than others, not very frequent buyer/visitor and generates very low monetary. There are 1821 customers (26%) within the cluster
- Mid Value: In the middle of everything. Often buying our product (but not as much as our High Values), fairly frequent and generates moderate monetary. There are 2833 customers (41%) within the cluster
- High Value: The group we don’t want to lose. High Monetary, Frequency and low Inactivity. There are 2235 customers (33%) within the cluster

Based on the clustering result, it's recommended to marketing team to start taking action with this main strategies:
- High Value: Create loyalty program
- Mid Value: Increase frequency
- Low Value: Improve retention + Increase frequency
